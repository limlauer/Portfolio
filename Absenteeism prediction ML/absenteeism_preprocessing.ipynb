This is a categorical nominal variable (reason).

### Getting the dummies

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True) # drop_first to drop the reason for absence column (avoiding multicolinearity problems)
reason_columns = reason_columns.astype(int) # If we don't do this, the final output will give us True and False instead of 1 and 0
print(reason_columns)
reason_columns['check'] = reason_columns.sum(axis=1) # Check there is a reason for every row
print(f"Total: {reason_columns['check'].sum()}")
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1) # All good, we can drop the check column
```

### Age
Let's explore the ages now

```python
age_values_sorted = sorted(raw_csv['Age'].unique())
age_values_row = ', '.join(map(str, age_values_sorted))

print(age_values_row)
```

27 to 58 years old
No need to change anything here, lets keep going.

### Grouping the reasons for absence
We can't have 28 dummies, let's group them in 4 categories and concatenate them to our dataframe, checking the data as we go

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
df
```

#### Groups (according to the feature descriptions table)

- Group 1: diseases | 1 to 14
- Group 2: pregnancy and giving birth | 15 to 17
- Group 3: poisoning | 18 to 21
- Group 4: light reasons for absence (medical appointments) | 22 to 28

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis = 1)
print(reason_type_1)
```

```python
df
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
df
```

Perfect, now we will change the order of the columns to match the original format

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']

df.columns = column_names
df.head()
```

```python
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_names_reordered]
df.head()
```

We are done with the reasons for absence, let's make a copy of the dataframe and continue

### Dates
We need to check the dates format and change it if necessary

```python
df2 = df.copy()
```

```python
df2.dtypes
```

```python
# 23/07/2015 <- dd/mm/yyyy
df2['Date'] = pd.to_datetime(df2['Date'], format = '%d/%m/%Y')
```

```python
print(df2.info())
```

### Extracting the month from the date
We are going to separate the month and weekday from the date, they might give us important information for our ML model

```python
list_months = []
df_length = df2.shape[0]
for i in range(df_length):
    list_months.append(df2['Date'][i].month)

list_months
```

```python
df2['Month Value'] = list_months
df2.head()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

### Extracting the weekday
Let's use .map() and a function to get the day of the week of a given date

```python
df2['Day of the week'] = df2['Date'].apply(date_to_weekday)
```

```python
df2
```

```python
df2 = df2.drop('Date', axis = 1)
df2.columns
```

```python
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month Value', 'Day of the week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']

df2 = df2[column_names_reordered]
df2.head()
```

```python
df3 = df2.copy()
df3
```

### Transforming the education column
INFO:
- 1 high school
- 2 graduate
- 3 postgraduate
- 4 master or doctor

Let's change every value from high school to 0 and the others to 1 

```python
print(df3['Education'].unique())
print(df3['Education'].value_counts())
```

```python
df3['Education'] = df3['Education'].map({
    1:0,
    2:1,
    3:1,
    4:1
})
print(df3['Education'].unique())
print(df3['Education'].value_counts())
```
